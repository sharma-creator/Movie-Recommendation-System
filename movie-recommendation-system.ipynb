```python
# function to extract the genres name from each genres row
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert1(obj):
    L = []
    data = [item['name'] for item in ast.literal_eval(obj)[0:3]]
    L.append(data)
    return L
```

```python
[item['name'] for item in list1[:3]]
```

```python
def convert_cast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['cast'][0][1].replace(' ','')
```

```python
def space(obj):
    L = []
    for i in obj:
        L.append(i.replace(' ',''))
    return L
```

```python
movies['cast'] = movies['cast'].apply(space)
movies['genres'] = movies['genres'].apply(space)
movies['keywords'] = movies['keywords'].apply(space)
movies['crew'] = movies['crew'].apply(space)
```

```python
movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tag']]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
# Vectorization
# Bag of Words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stems(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(stems)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
stems(new_df['tag'][0])
```

```python
# Cosine to find similarity b/w two vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# Sorted index of similarity
sorted(similarity[0],reverse=True) # The problem with this that we dont have the index of the movie here
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:11]
```

```python
print(type(list))
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
